```python
import numpy as np
import matplotlib.pyplot as plt

# Matrix
M = np.array([[-np.sqrt(3), 3], [5, -np.sqrt(3)]])/2.
# SVD of M
U, S, VT = np.linalg.svd(M)
D = np.diag(S)

# Vector
unit_x = np.array([1, 0])
unit_y = np.array([0, 1])
```

```python
# Apply SVD step by step
# Plot vectors
plt.quiver(0, 0, *unit_x, color='r', angles='xy', scale_units='xy', scale=1, alpha=0.4)
plt.quiver(0, 0, *unit_y, color='b', angles='xy', scale_units='xy', scale=1, alpha=0.4)

# First apply V^T...
unit_x1 = VT.dot(unit_x)
unit_y1 = VT.dot(unit_y)
plt.quiver(0, 0, *unit_x1, color='r', angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, *unit_y1, color='b', angles='xy', scale_units='xy', scale=1)

theta = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(theta), np.sin(theta), 'k--')
plt.axis('equal')
plt.savefig('svd_v.pdf')
```

```python
# Apply D...
unit_x2 = D.dot(unit_x1)
unit_y2 = D.dot(unit_y1)

plt.quiver(0, 0, *unit_x1, color='r', angles='xy', scale_units='xy', scale=1, alpha=0.4)
plt.quiver(0, 0, *unit_y1, color='b', angles='xy', scale_units='xy', scale=1, alpha=0.4)

plt.quiver(0, 0, *unit_x2, color='r', angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, *unit_y2, color='b', angles='xy', scale_units='xy', scale=1)

plt.plot(np.cos(theta), np.sin(theta), 'k--')
plt.axis('equal')
plt.axis([-2.6, 2.6, -2.6, 2.6])
plt.savefig('svd_d.pdf')
```

```python
# Apply U...
unit_x3 = U.dot(unit_x2)
unit_y3 = U.dot(unit_y2)

plt.quiver(0, 0, *unit_x2, color='r', angles='xy', scale_units='xy', scale=1, alpha=0.4)
plt.quiver(0, 0, *unit_y2, color='b', angles='xy', scale_units='xy', scale=1, alpha=0.4)

plt.quiver(0, 0, *unit_x3, color='r', angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, *unit_y3, color='b', angles='xy', scale_units='xy', scale=1)

theta = np.linspace(0, 2*np.pi, 100)
plt.plot(np.linalg.norm(unit_x3)*np.cos(theta), np.linalg.norm(unit_x3)*np.sin(theta), 'k--')
plt.plot(np.linalg.norm(unit_y3)*np.cos(theta), np.linalg.norm(unit_y3)*np.sin(theta), 'k--')
plt.axis('equal')
plt.savefig('svd_u.pdf')
```
