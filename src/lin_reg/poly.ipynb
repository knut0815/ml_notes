```python
import numpy as np
import matplotlib.pylab as plt

# Number of samples
m = 50
# Create samples of house size in 100 ft^2
min_x = 4
max_x = 30
x = np.random.uniform(min_x, max_x, m)
# Create house prices ($) in 1000's, by assuming the
# relationship y = theta0 + theta1 * x + theta2 * x^3 + eps,
# where eps is noise.
theta0 = 300
theta1 = -2
theta2 = 0.05
y = theta0 + theta1 * x + theta2 * x**3 + np.random.uniform(-100, 100, m)
```

```python
def design_mat(x):
    return np.column_stack((np.repeat(1, len(x)), x, x**2, x**3))

# Create design matrix
X = design_mat(x)
# Use normal equation to estimate thetas
thetas = np.linalg.inv(np.transpose(X).dot(X)).dot(np.transpose(X)).dot(y)
```

```python
# Predit house prices
sizes = np.linspace(min_x, max_x, 100)
prices = design_mat(sizes).dot(thetas)
# Plot data
plt.plot(x, y, 'k.')
# Plot fit
plt.plot(sizes, prices, 'r-')
plt.xlabel("House size (100 sq ft)", fontsize=11)
plt.ylabel("House price ($1000)", fontsize=11)
plt.savefig("poly-data.pdf")
```
