```python
import numpy as np
import matplotlib.pylab as plt
```

```python
# Number of samples
m = 20
# Create samples of house size in ft^2
min_x = 400
max_x = 3000
x = np.random.uniform(min_x, max_x, m)
# Create house prices ($) in 1000's, by assuming the
# relationship y = theta * x (1 + \eps), where \eps
# is noise.
optimal_theta = 0.3
y = optimal_theta * x * (1 + np.random.uniform(-0.3, 0.3, m))
```

```python
cost_func = lambda theta: 1/(2*m) + np.sum(((theta * x) - y)**2)
```

```python
# Different values of theta and their corresponding
# colors for plotting
thetas = [(0, 'r'), (optimal_theta, 'g'), (0.5, 'b')]

# Put a bit of space between the subplots
plt.subplots_adjust(wspace=0.3)

# Plot data and various fits to data
plt.subplot(121)
# Plot training set
plt.plot(x, y, 'k.')
# Plot three lines, one with each value of theta
x_vals = np.linspace(min_x, max_x, 100)
for (theta, color) in thetas:
    plt.plot(x_vals, theta * x_vals, color)
plt.xlabel(r"$x$", fontsize='12')
plt.ylabel(r"$y$", fontsize='12')

# Plot cost function
plt.subplot(122)
# Plot cost function curve
theta_vals = np.linspace(-0.5, 1, 100)
plt.plot(theta_vals, [cost_func(t) for t in theta_vals], 'k-')
# Plot cost function on specific thetas
for (theta, color) in thetas:
    plt.plot(theta, cost_func(theta), "{}p".format(color), markersize=10)

# Save plot
plt.xlabel(r'$\theta_1$', fontsize='12')
plt.ylabel(r"$J$", fontsize='12')
plt.savefig("costfunc.pdf")
```
